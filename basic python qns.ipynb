Given a list, write a Python program to swap first and last element of the list.

Examples:

Input : [12, 35, 9, 56, 24]
Output : [24, 35, 9, 56, 12]

Input : [1, 2, 3]
Output : [3, 2, 1]

```python
list_input = [int(i) for i in input().split(' ')]

#removing last elament and taking into a valriable
last_ele = list_input.pop()

#removing first elament and taking into a valriable
frist_ele = list_input.pop(0)

#inserting last element to first position
list_input.insert(0, last_ele)

#inserting first element to last position
list_input.append(frist_ele)

list_input
```

 Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings. \
Sample List : ['abc', 'xyz', 'aba', '1221']\
Expected Result : 2

```python
input_list = ['abc', 'xyz', 'aba', '1221']
count = 0
for i in input_list:
    if len(i) >= 2:
        if i[0] == i[-1]:
            count += 1
print(count)
```

```python
Sample_List = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
new = [(0,0)]
a = True
while a:
    for i in Sample_List:
        for k,j in enumerate(new):
            if i[-1] <= j[-1]:
                new.insert(k, i)
                break
        else:
            new.append(i)
    break
new.pop(0)
new
```

n =5 \

```python
Sample_List = ['p', 'q']
n = 5
new = []
for i in range(1,n+1):
    for j in Sample_List:
        new.append(str(j)+str(i))
new
```

```python
#HCF/GCD
def hcf(a, b):
    value = int()
    if a > b:
        pass
    else:
        a, b = b, a
    for i in range(1,a):
        if a % i == 0:
            if b % i == 0:
                value = i
                
    return value
z = hcf(12, 30)   
z
```
