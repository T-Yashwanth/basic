# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df_clean = pd.DataFrame(people)
df_clean
```

```python
df_clean.info()
```

```python
df_clean.dropna()
```

```python
#(how) is used to filter the drop values
df_clean.dropna(axis = 'index', how='all')
```

```python
df_clean.dropna(axis = 'index', how='any')
```

```python
df_clean.dropna(axis = 'columns', how='any')
```

```python
#(subset) is used to keep the required fields it it is not empty
df_clean.dropna(axis = 'index', how='all', subset=['email','last'])
```

```python
df_clean
```

```python
df_clean.isna()
```

```python
df_clean.replace("NA", np.nan,inplace=True)
df_clean.replace("Missing", np.nan,inplace=True)
df_clean
```

```python
df_clean.isna()
```

```python
#to fill data in the place of missing values
df_clean.fillna('missing')
```

```python
df_clean.dtypes
```

```python
#age has a NAN values so it is a string we can convert it into a float but not integer
df_clean['age'] = df_clean['age'].astype(float)
```

```python
df_clean['age'].mean()
```
